```python
class context(object):
 
    def __init__(self, question, begin, end, answer):
        self.question = question
        self.begin = begin
        self.end = end
        self.answer = answer
    def print_c(self):
        print(" ".join(self.question), "(", self.begin, self.end, ")")
    def print_v(self):
        print(self.question, "\n(", self.begin, self.end, ")")
```

```python
def tockenize (vector):
    from nltk.tokenize import RegexpTokenizer
    #tokenizer = RegexpTokenizer('\w+|\$[\d]+|\S+')
    tokenizer = RegexpTokenizer('\w+') 
    #tokenizer = RegexpTokenizer('(\w+([\'-]\w+)*)')
    tokens = tokenizer.tokenize(vector)
    words = [w.lower() for w in tokens if w.isalnum()]
    return words
```

```python
def get_emb(size):
    m = len(paragraphs_ft)
    l = np.arange(0, m, 1)
    np.random.shuffle(l)
    l = l[:size]
    emb = np.zeros((size, CONTEXT_MAX_SIZE, EMBEDDING_SIZE))
#     mask = np.zeros((size, CONTEXT_MAX_SIZE))
    len_emb = np.zeros((size), dtype=np.int32)
    for i in range(size):
        count = 0
        for j in range(CONTEXT_MAX_SIZE):
            emb[i][j] = (model_s[paragraphs_ft[l[i]][j]])
#             if paragraphs_ft[l[i]][j] != 0 and paragraphs_ft[l[i]][j] != 1:
#                 mask[i][j] = 1
            if paragraphs_ft[l[i]][j] != 1:
                count += 1
        len_emb[i] = count + 1
#         mask[i][count] = 1
#     print(len_emb)
    return emb, l, len_emb
def get_answ(size, l):
    #start = np.zeros((size, CONTEXT_MAX_SIZE))
    start = np.zeros((size))
    emb = np.zeros((size, QUESTION_MAX_SIZE, EMBEDDING_SIZE))
    for i in range(size):
        q = 0
        if (0 < len(features_ft[l[i]])):
            q = np.random.randint(0, len(features_ft[l[i]]), 1)
            q = q[0]
            #q = 0
            for j in range(QUESTION_MAX_SIZE):
                emb[i][j] = (model_s[features_ft[l[i]][q].question[j]])
            start[i] = (features_ft[l[i]][q].begin)
            #start[i] = (features_ft[l[i]][q].begin)
    return emb, start
```
